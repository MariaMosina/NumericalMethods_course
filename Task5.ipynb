# $f(x) = tg(0.5x+0.2) - x^2$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a, b = -2.5, 2.5
```

```python
def Lagrange(x, xs, fs):
    res = 0
    for i in range(len(xs)):
        num, den = 1, 1
        for x_j in xs[:i]+xs[i+1:]:
            num *= x - x_j
            den *= xs[i] - x_j
        res += (num/den) * fs[i]
    return res
```

```python
def points(start, end, n , f):
    xs = np.linspace(start, end, n)
    fs = list(f(xs))
    xs = list(xs)
    xs_opt = []
    for i, x in enumerate(xs):
        xs_opt.append(0.5 * ((end-start)*np.cos(((2*i+1)*np.pi)/(2*n+2)) + (end+start)))
    fs_opt = list(f(np.array(xs_opt)))
    xs_opt = list(xs_opt)
    return xs, fs, xs_opt, fs_opt
```

```python
f = lambda x: np.tan(0.5*x+0.2) - x**2
fig, ax = plt.subplots()
ax.set_title('$ tg(0.5x+0.2) - x^2$')
x = np.linspace(a, b, 100)
plt.plot(x, f(x))
plt.show()
```

```python
xs, fs, xs_opt, fs_opt = points(a, b, 6, f)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
fig.set_figheight(5)
fig.set_figwidth(15)
ax1.set_title('x')
ax2.set_title('delta x')

x = np.linspace(a, b, 100)
ax1.plot(x, f(x))
f_xs = []
for xx in x:
    f_xs.append(Lagrange(xx, xs, fs))
f_xs = np.array(f_xs)
ax1.plot(x, f_xs)
ax2.plot(x, abs(f_xs-f(x)))

fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
fig.set_figheight(5)
fig.set_figwidth(15)
ax1.set_title('x opt')
ax2.set_title('deltax opt')

x = np.linspace(a, b, 100)
ax1.plot(x, f(x))
f_xs = []
for xx in x:
    f_xs.append(Lagrange(xx, xs_opt, fs_opt))
f_xs = np.array(f_xs)
ax1.plot(x, f_xs)
ax2.plot(x, abs(f_xs-f(x)))
```

```python
def Newton(xs, fs):
    x = Symbol('x')
    res = fs[0]
    n = len(xs)
    for k in range(2, n+1):
        
        fi = 0
        for i in range(k):
            g = 1
            for xi in xs[:i]+xs[i+1:k]:
                g *= xs[i] - xi
#             print('g', g)
            g = 1/g
            fi += g*fs[i]
            
        dot = 1
        for i in range(k-1):
            dot *= x-xs[i]
        res += fi * dot
#         print(k, fi, dot)
    return simplify(res)
```
